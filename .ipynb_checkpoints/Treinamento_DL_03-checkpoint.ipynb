O TensorFlow é uma estrutura de código aberto desenvolvida por pesquisadores do Google para executar aprendizado de máquina, aprendizado profundo e outras cargas de trabalho de análise estatística e preditiva. Como plataformas semelhantes, ele foi projetado para agilizar o processo de desenvolvimento e execução de aplicativos analíticos avançados para usuários como cientistas de dados, estatísticos e modeladores preditivos.

O software TensorFlow lida com conjuntos de dados que são organizados como nós computacionais em forma de gráfico. As arestas que conectam os nós em um grafo podem representar vetores ou matrizes multidimensionais, criando o que chamamos de tensores. Como os programas do TensorFlow usam uma arquitetura de fluxo de dados que funciona com resultados intermediários generalizados dos cálculos, eles são especialmente abertos a aplicativos de processamento paralelo de grande escala, sendo as redes neurais um exemplo comum.

A estrutura inclui conjuntos de APIs de alto e baixo nível. O Google recomenda usar os de alto nível quando possível para simplificar o desenvolvimento do pipeline de dados e a programação de aplicativos. No entanto, saber como usar as APIs de baixo nível - chamadas TensorFlow Core - pode ser valioso para experimentação e depuração de aplicativos, diz a empresa; também dá aos usuários um "modelo mental" do funcionamento interno da tecnologia de aprendizado de máquina, nas palavras do Google.

### O que é Tensor no Tensorflow



TensorFlow, como o nome indica, é uma estrutura para definir e executar cálculos envolvendo tensores. Um tensor é uma generalização de vetores e matrizes para dimensões potencialmente mais altas. Internamente, o TensorFlow representa tensores como matrizes n-dimensionais de tipos de dados básicos. Cada elemento no Tensor tem o mesmo tipo de dados, e o tipo de dados é sempre conhecido. A forma (ou seja, o número de dimensões que possui e o tamanho de cada dimensão) pode ser apenas parcialmente conhecida. A maioria das operações produz tensores de formas totalmente conhecidas se as formas de suas entradas também forem totalmente conhecidas, mas em alguns casos só é possível encontrar a forma de um tensor no tempo de execução do gráfico.

```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
from matplotlib import cm
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib import cm
from matplotlib.colors import ListedColormap
import geopandas as gpd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
samples = gpd.read_file('/content/drive/MyDrive/Datasets/LULC_Pixel_Classification/classes_de_uso.shp')
```

```python
samples
```

```python
samples.columns
```

```python
X = samples[['B12', 'B8A', 'B2', 'B3', 'B4', 'B5', 'B6', 'B7', 'B8']].values
Y = samples['classe'].values
```

```python
Y = Y[:,np.newaxis]
```

```python
Y.shape
```

```python
enc = OneHotEncoder()

enc.fit(Y)

Y = enc.transform(Y).toarray()
```

```python
Y.shape
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)
```

Agora vamos definir o tamanho do shape de entrada e o número de classes que temos:

```python
input_shape = (X_train.shape[1:])
num_classes = len(np.unique(samples['classe'].values))
```

```python
input_shape
```

```python
num_classes
```

Então vamos construir nossa rede neural adicionando as camadas densas:

```python
model = Sequential()
model.add(Dense(128, input_shape=input_shape, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))
```

Na camada de entrada e nas camadas ocultas usamos a função de ativação relu. Como estamos lidando com um problema de classificação multiclasse, usaremos a função softmax enable na camada de saída.

Agora podemos compilar o modelo, definindo a perda como entropia categórica, o otimizador Adam e a métrica de validação como precisão:

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

E, finalmente, treine o modelo para 100 iterações:

```python
history = model.fit(X_train, Y_train, epochs=300, batch_size=250, verbose=1, validation_split=0.2)
```

Vamos plotar os gráficos de loss e acurácia por épocas para os dados de treinamento e validação:

```python
fig, ax = plt.subplots(1,2, figsize=(16,8))
ax[0].plot(history.history['loss'], color='b', label="Training loss")
ax[0].plot(history.history['val_loss'], color='r', label="validation loss",axes =ax[0])
legend = ax[0].legend(loc='best', shadow=True)

ax[1].plot(history.history['accuracy'], color='b', label="Training accuracy")
ax[1].plot(history.history['val_accuracy'], color='r',label="Validation accuracy")
legend = ax[1].legend(loc='best', shadow=True)
```

```python
score = model.evaluate(X_test, Y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred_res = np.argmax(y_pred, axis=1)
```

```python
Y_test_res = np.argmax(Y_test, axis=1)
```

```python
print(classification_report(Y_test_res, y_pred_res))
```

#Dropout

Dropout é um método de regularização que aproxima o treinamento de um grande número de redes neurais com diferentes arquiteturas em paralelo.

Durante o treinamento, algumas saídas da camada são ignoradas aleatoriamente ou "retiradas". Isso tem o efeito de fazer com que a camada pareça e seja tratada como uma camada com um número diferente de nós e conectividade com a camada anterior. Com efeito, cada atualização de uma camada durante o treinamento é realizada com uma “visualização” diferente da camada configurada.

image.png

## Batch Normalization

Antes de entrar na normalização de lote, vamos entender o termo “Normalização”.

A normalização é uma ferramenta de pré-processamento de dados usada para trazer os dados numéricos para uma escala comum sem distorcer sua forma.

Geralmente, quando inserimos os dados em uma máquina ou algoritmo de aprendizado profundo, tendemos a alterar os valores para uma escala equilibrada. A razão pela qual normalizamos é, em parte, para garantir que nosso modelo possa generalizar adequadamente.

Agora, voltando à normalização em lote, é um processo para tornar as redes neurais mais rápidas e estáveis por meio da adição de camadas extras em uma rede neural profunda. A nova camada executa as operações de padronização e normalização na entrada de uma camada proveniente de uma camada anterior.

Mas qual é a razão por trás do termo “Lote” na normalização de lote? Uma rede neural típica é treinada usando um conjunto coletado de dados de entrada chamado lote. Da mesma forma, o processo de normalização na normalização em lote ocorre em lotes, não como uma única entrada.

```python
from tensorflow.keras.layers import Dropout, BatchNormalization
```

Vamos então adicionar uma camada de dropout e batch normalization entre as camadas densas e ver como o modelo se comporta:

```python
model_2 = Sequential()
model_2.add(Dense(128, input_shape=input_shape, activation='relu'))
model_2.add(Dropout(0.1))
model_2.add(BatchNormalization())
model_2.add(Dense(32, activation='relu'))
model_2.add(Dropout(0.1))
model_2.add(Dense(8, activation='relu'))
model_2.add(Dropout(0.1))
model_2.add(Dense(num_classes, activation='softmax'))
```

```python
model_2.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model_2.summary()
```

Vamos treinar o modelo:

```python
history2 = model_2.fit(X_train, Y_train, epochs=300, batch_size=250, verbose=1, validation_split=0.2)
```

Em seguida, vamos plotar as curvas de loss e acurácia:

```python
fig, ax = plt.subplots(1,2, figsize=(16,8))
ax[0].plot(history2.history['loss'], color='b', label="Training loss")
ax[0].plot(history2.history['val_loss'], color='r', label="validation loss",axes =ax[0])
legend = ax[0].legend(loc='best', shadow=True)

ax[1].plot(history2.history['accuracy'], color='b', label="Training accuracy")
ax[1].plot(history2.history['val_accuracy'], color='r',label="Validation accuracy")
legend = ax[1].legend(loc='best', shadow=True)
```

```python
y_pred_2 = model_2.predict(X_test)
```

```python
y_pred_res2 = np.argmax(y_pred_2, axis=1)
```

```python
Y_test_res2 = np.argmax(Y_test, axis=1)
```

```python
print(classification_report(y_pred_res2, Y_test_res2))
```

```python
c_matrix = confusion_matrix(Y_test_res2, y_pred_res2)
```

```python
names = ['Agua','Vegetação','Solo','Agricultura','Infra_urbana']
```

```python
r1 = pd.DataFrame(data=c_matrix, index= names, columns=names)
fig, ax = plt.subplots(figsize=(8,8))
ax = sns.heatmap(r1, annot=True, annot_kws={"size": 18},fmt='d',cmap="Blues", cbar = False)
#for t in ax.texts: t.set_text(t.get_text() + " %")
ax.tick_params(labelsize=16)
ax.set_yticklabels(names, rotation=45)
ax.set_ylabel('True')
ax.set_xlabel('Predict')
```
